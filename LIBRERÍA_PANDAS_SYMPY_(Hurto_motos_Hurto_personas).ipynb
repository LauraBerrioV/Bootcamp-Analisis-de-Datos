```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris["target"] = iris.target
df_iris
```

```python
df_iris.columns
```

```python
#Renombrar la columna target por species
df_iris.rename({"target": "species"}, inplace=True, axis=1)
df_iris
```

```python
#Transformo el tipo de dato del campo species, a dato categórico
df_iris["species"]= df_iris["species"].astype("category")
df_iris.info()
```

```python
#Asignar valores  nulos a la columna species
#perc=0.25 significa que voy a asignar valores nulos al 25% de los datos
iris_amp = mf.ampute_data(df_iris, perc=0.25, random_state=1991) # Realizo una amputación de datos aplicando NaN, de forma aleatoria
iris_amp
```

```python
# Crear kernel
kds = mf.ImputationKernel(
    iris_amp, # Nombre de dataframe para entrenar modeolo
    random_state=1991  #Valores aleatorios para entrenar modelo
)

# Corre el algoritmo MICE durante dos iteraciones
kds.mice(2)

# Retorna el dataset completo
iris_completo = kds.complete_data()
iris_completo
```

```python
## Calcular la autocorrelación de la serie de tiempo

# Para entender la correlacion, tomar el ejemplo de que estoy analizando los robos de hoy martes, y los comparo con los robos de martes de la semana pasada, si coinciden en la cantidad hay correlacuion , entonces ya se que debo evitar ese dia para evitar un robo
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(12,6))
plot_acf(df_fecha_agr) # Se puede ajustar el número de lags que se necesitan . LAg es un retraso o periodo de tiempo hacia atras con el que se hace la correlacion
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Autocorrelation Function (ACF) Plot")
plt.show()
```
